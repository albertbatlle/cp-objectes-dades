```python
import matplotlib.pyplot as plt
import seaborn as sns

# sns.get_dataset_names()
pinguins = sns.load_dataset("penguins")
pinguins.columns = ["especie", "illa", "llargada_bec", "profunditat_bec", "llargada_aleta", "massa", "sexe"]
pinguins
```

## Gràfica relacional

### Estils
style="white", "dark", "whitegrid", "darkgrid", "ticks" </br>
context="paper", "notebook", "talk", "poster"

```python
sns.set_theme(style="darkgrid", context="talk")
# 
relational = sns.relplot(data=pinguins, height=7, aspect=1.2, markers=["s", ".", "P", "d"])
relational.set(title="Anàlisis de pinguins")
relational.set_ylabels("Variables")
relational.set_xlabels("Número de pinguins")
# plt.grid(False) # treu el grid
plt.show()
```

### Diagrames de dispersió

```python
print(">>>> SIMPLE")

sns.relplot(data=pinguins, x="llargada_bec", y="llargada_aleta")

print(">>>> MATIZ I TAMANY FIXE")

sns.relplot(data=pinguins, x="llargada_bec", y="llargada_aleta", hue="massa", s=300)

print(">>>> REJILLA, TAMANY PERSONALITZAT I ESTIL")
```
