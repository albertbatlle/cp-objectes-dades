```python
# Importar llibreries
import numpy as np
from matplotlib import pyplot as plt
```

```python
print(df6["region"][:9])
```

```python
chicago = df6[df6["region"] == "Chicago"]
print(chicago.head(6))
#chicago
```

```python
#ordenar per data i per això canviem l'índex
chicago = chicago.set_index("Date")
chicago = chicago.sort_values(by="Date")

print(chicago.head(6))
```

```python
#Gràfic
max_sample = 100
precio = chicago["AveragePrice"][: max_sample]
cantidad = chicago["Total Volume"][: max_sample]

plt.plot(precio, label="Precio medio")
plt.plot(cantidad, label="Volumen total")

plt.title("Precio aguacates vs tiempo")
plt.xlabel("Fecha")
plt.xticks(rotation=90)
plt.ylabel("Precio en $")
plt.legend()
plt.show()
```

```python
atlanta = df6[df6["region"] == "Atlanta" ]
precio = atlanta["AveragePrice"]
volumen = atlanta["Total Volume"]

bolsasAguacate = atlanta["Total Bags"]
sAguacate = atlanta["Small Bags"]
lAguacate = atlanta["Large Bags"]
xAguacate = atlanta["XLarge Bags"]

# els número de baix són 2 files 2 columnes i el 1r eix
plt.subplot(221)
plt.title("Precio aguacate")
plt.plot(precio, label="Precio", color="green")
plt.legend()

plt.subplot(222)
plt.title("Volumen aguacates")
plt.plot(volumen, label="Volumen Total", color="red")
plt.legend()

plt.subplot(223)
plt.title("Bolsas Totales")
plt.plot(bolsasAguacate, label="Bolsas Totales", color="blue")
plt.legend()

plt.subplot(223)
plt.title("Bolsas por tamaño")
plt.plot(sAguacate, label="Bolsas - S", color="black")
plt.plot(lAguacate, label="Bolsas - L", color="cyan")
plt.plot(xAguacate, label="Bolsas - XL", color="yellow")
plt.legend()

plt.show()
```
